## Набір даних Taxis

The trip records for yellow and green taxis as well as for-hire vehicles (FHVs) include essential fields such as pick-up and drop-off dates/times, locations, trip distances, fares, rate types, payment methods, and passenger counts. These datasets were collected and provided to the NYC Taxi and Limousine Commission (TLC) by authorized technology providers under the Taxicab & Livery Passenger Enhancement Programs (TPEP/LPEP) and FHV Trip Record submissions made by bases. However, the TLC does not create this trip data and does not guarantee its accuracy.  [link](https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page).  

Які колонки мість набір даних:  
pickup - trip start time
dropoff - trip end time
passengers - the number of passengers on the trip
distance - the distance of the trip
fare - the basic fare for the trip
tip - tip size
tolls - payment for toll roads (the passenger also pays for this)
total - total fare for the trip
color - the color of the car that made the trip
payment - payment method
pickup_zone - the zone in which the trip began
dropoff_zone - the zone in which the trip ended
pickup_borough - the district where the trip started
dropoff_borough - the district where the trip ended 
  
You can view all data sets from seaborn [here](https://github.com/mwaskom/seaborn-data).


```python
df_taxi = sns.load_dataset('taxis')
```

Let's check if there are any missing values in this data set.

```python
df_taxi.isna().sum()
```

Why there are missing data in several columns in these rows?

```python
rows_with_na_dropoff_zone = df_taxi[df_taxi['dropoff_zone'].isna()]
display(rows_with_na_dropoff_zone)
```

1. There are missing data in the columns with the place of disembarkation of the passenger and sometimes with the place of boarding. Perhaps this is due to the absence of a GPS signal or communication in that place at that time and, accordingly, the data was entered incorrectly in the plate. 2. There are lines where the distance has a zero value and the payment amount is a non-zero value. I noticed that in most of these records the trip end time is the same or a few seconds longer than the trip start time. Also, there are often no data on both the place of landing and the place of disembarkation. Most likely, the trip took place, funds from the client were credited, but there were communication failures and the data was incorrectly recorded

```python
df_taxi_clean = df_taxi.fillna('NA')
```

```python
df_taxi_clean.isna().sum()
```

```python
df_taxi_clean['pickup_day'] = df_taxi_clean.pickup.dt.floor('d')
```

```python
df_taxi_clean_grouped_by_day = df_taxi_clean.groupby('pickup_day')['total'].sum()
df_taxi_clean_grouped_by_day.plot()
plt.xlabel('Date')
plt.ylabel('Payments')
plt.title('Payments per day')
plt.grid(True)
plt.show()
```

```python
There is a very low value at the beginning of the graph
```

```python
pickup_day_count = df_taxi_clean['pickup_day'].value_counts()

display(pickup_day_count)
```

```python
extra_date = df_taxi_clean[df_taxi_clean['pickup_day'] == pd.to_datetime('2019-02-28')].index

df_taxi_clean = df_taxi_clean.drop(extra_date)

df_taxi_clean_grouped_by_day = df_taxi_clean.groupby('pickup_day')['total'].sum()
df_taxi_clean_grouped_by_day.plot()
plt.xlabel('Date')
plt.ylabel('Payments')
plt.title('Payments per day')
plt.grid(True)
plt.show()
```

# Let's look at the dynamics of total payments by borough.

```python
borough_pivot = df_taxi_clean.pivot_table(values='total', index='pickup_day', columns='pickup_borough', aggfunc='sum')
borough_pivot = borough_pivot.drop(columns='NA')

print(borough_pivot)
```

# In which borough is the highest amount of daily payments for taxi trips?

```python
borough_pivot.fillna(0).plot();
borough_pivot.plot()
```

The highest amount of daily payments for taxi trips is in Bronx')

# Let's analyze the sums of daily payments for trips in each of the districts

```python
borough_pivot.hist(bins=10)
```

# Let's compare the distribution of daily sums of payments by boroughs.

```python
x1 = borough_pivot['Bronx']
x2 = borough_pivot['Brooklyn']
x3 = borough_pivot['Manhattan']
x4 = borough_pivot['Queens']

plt.hist(x1, alpha = 0.3, label= 'Bronx')
plt.hist(x2, alpha = 0.3, label= 'Brooklyn')
plt.hist(x3, alpha = 0.3, label= 'Manhattan')
plt.hist(x4, alpha = 0.3, label= 'Queens')

plt.title('Daily total distributions by boroughs')
plt.xlabel('Daily total')
plt.ylabel('Count of ride')

plt.legend()

plt.show()
```

# Is there a relationship between the distance of the trip and the size of the tip?

# Let's analyze the dynamics of tips in the boroughs where the passenger was picked up.

```python
tips_pivot = df_taxi_clean.pivot_table(values='tip', index='pickup_day', columns='pickup_borough', aggfunc='mean').drop(columns='NA')
tips_pivot.fillna(0).plot();
plt.xlabel('Pickup Days')
plt.ylabel('Tip')
```

```python
The lowest average amount of tips is while picking-up in Bronx
The highest average amount of tips is while picking-up in Queens
The most stable average amount of tips is while picking-up in Manhattan
```

# Let's build a box-plot of tip size by district of trip start

```python
plt.figure(figsize=(10, 5))

sns.boxplot(data=df_taxi_clean[df_taxi_clean['pickup_borough'] != 'NA'], x='pickup_borough', y='tip')
plt.xlabel('Pickup Borough')
plt.ylabel('Tip')
plt.title('Tip by Pickup Borough')

plt.show()
```

# Does the number of passengers affect the average tip?

The number of passengers does not affect the average tip size

# On which days are the largest payments for trips?

```python
df_taxi_clean['day_of_week'] = pd.to_datetime(df_taxi_clean['pickup']).dt.day_name()
df_taxi_clean_grouped_by_day_of_week = df_taxi_clean.groupby('day_of_week')['total'].sum()
df_taxi_clean_grouped_by_day_of_week.plot()
plt.xlabel('Day')
plt.ylabel('Payments')
plt.title('Payments per day')
plt.grid(True)
plt.show()
```

The highest amounts for trips on Fridays, the lowest on Mondays. This should be taken into account when forming a marketing strategy and driver engagement

# At what hours are the largest payments for trips?

```python
df_taxi_clean['hour'] = df_taxi_clean['pickup'].dt.hour

df_taxi_clean_grouped_by_hour = df_taxi_clean.groupby('hour')['total'].sum()

df_taxi_clean_grouped_by_hour.plot()
plt.xlabel('Hour')
plt.ylabel('Payments')
plt.title('Payments per hour')
plt.grid(True)

plt.show()
```

The largest payments for trips are received between 3 and 8 p.m., so there should be enough drivers working during these hours.
